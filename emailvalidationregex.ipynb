## Email Validation

```python
import re
def validation_of_email(email):
    pattern1=r"^[a-z0-9._%+-]+@[a-z]+\.[a-zA-Z.]+$"
    # pattern2=r"^[A-Z0-9._%+-]+@[A-Z]+\.[A-Z]+$"
    if re.match(pattern1,email):
        return True
    return False
valid_emails = [
    "user@gmail.com",
    "firstname.lastname@gmail.com",
    "user_name123@yahoo.com",
    "user+promo@gmail.com",
    "contact@company.co.in",
    "info@mail-domain.org",
    "student123@iitb.ac.in",
    "admin@my-site.net",
    "test.email@sub.domain.com",
    "a.b.c.d@gmail.com"
]
for email in valid_emails:
    print(email,"->",validation_of_email(email)) 
    

```
