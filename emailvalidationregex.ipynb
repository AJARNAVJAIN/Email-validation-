## Email Validation

```python
import re
def validation_of_email(email):
    pattern1=r"^[a-z0-9._%+-]+@[a-z]+\.[a-zA-Z.]+$"
    # pattern2=r"^[A-Z0-9._%+-]+@[A-Z]+\.[A-Z]+$"
    if re.match(pattern1,email):
        return True
    return False
email=input("Enter the Email to Verify. /n")
# for email in valid_emails:
print(email,"->",validation_of_email(email))
    

```
