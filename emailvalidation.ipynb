## Email Validation

```python
import re
def validation_of_email(email):
    pattern1=r"^[a-z0-9._%+-]+@[a-z]+\.[a-zA-Z.]+$"
    if re.match(pattern1,email):
        return "is a valid email"
    return "is not a valid email"
email="arnavjain@gmail.com" #input("Enter the Email to Verify. \n")
print(email,"->",validation_of_email(email))
    

```
